```python
import pandas as pd
import numpy as np

raw_df = pd.read_csv("Project Points.csv")
raw_df
```

```python

from numpy import isnan


raw_df['timestamp'] = pd.to_datetime(raw_df['timestamp'], errors='coerce')
raw_df
```

```python

lat_filter = (raw_df['lat']<90) & (raw_df['lat']>-90)
lon_filter = (raw_df['lon']<180) & (raw_df['lon']>-180)
timestamp_filter = (isnan(raw_df['timestamp']))
df = raw_df.loc[lat_filter & lon_filter & ~timestamp_filter]
df = df.sort_values(by=['timestamp']).reset_index(drop=True)
df
```

```python
removed_df = raw_df.loc[~(lat_filter & lon_filter & ~timestamp_filter)].reset_index(drop=True)
removed_df
```

removed_df.to_csv('rejects.log')

```python
import geopandas as gpd

gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat), crs='EPSG:4326')
gdf
```

```python
from shapely.geometry import LineString, Point
gdf['distance'] = 0
gdf
```

```python
gdf2 = gdf.shift()
gdf2
```

```python
from haversine import haversine
distance = gdf['geometry'].iloc[1:].combine(gdf2['geometry'].iloc[1:], haversine)
gdf['distance']=distance
gdf['distance'] = gdf['distance'].fillna(0)
gdf
```

```python
timespan = gdf['timestamp'].iloc[1:].sub(gdf2['timestamp'].iloc[1:])
gdf['timedelta'] = timespan.apply(lambda x: x.total_seconds()/60)
gdf = gdf.fillna(0)
gdf
```

```python

gdf['trip'] = ''
trip_start_index = 0
trip = 0
row = 0
while row < len(gdf):
    if gdf.loc[row, 'distance'] > 2 or gdf.loc[row, 'timedelta'] > 25:
        trip += 1
    gdf.loc[row, 'trip'] = trip
    row += 1
gdf.to_csv('with trip.csv')
```

```python
    
trips = list(gdf['trip'].unique())
trips_gdfs = {}

for trip in trips:
    trip_name = f'trip_{str(trip)}'
    per_trip_gdf = gdf.loc[gdf['trip'] == trip].reset_index(drop=True)
    trips_gdfs[trip_name] = per_trip_gdf
    # per_trip_gdf.to_csv(trip_name, index=False)

trips_jsons = {}

for trip in trips_gdfs:
    trip_gdf = trips_gdfs[trip]

    speed = trip_gdf.loc[1:, 'distance'].divide(trip_gdf.loc[1:, 'timedelta']) * 60
    trip_gdf['speed'] = speed

    max_speed = trip_gdf['speed'].max()
    total_km = trip_gdf.loc[1:,'distance'].sum()
    total_min = trip_gdf.loc[1:,'timedelta'].sum()
    average_speed = total_km/(total_min/60)

    trips_jsons[trip] = {
        'distance_km' : float(total_km),
        'duration_min': float(total_min),
        'avg_speed_kmh': float(average_speed),
        'max_speed_kmh': float(max_speed)
        }
    
trips_jsons
```

```python
from geojson import FeatureCollection, Feature, LineString
import random

feature_list = []
line_colors = []
for trip_name, trip_gdf in trips_gdfs.items():
    combine = list(zip(list(trip_gdf['lon']), list(trip_gdf['lat'])))

    random_color = ""
    while random_color not in line_colors:
        r = lambda: random.randint(0,255)
        random_color = '#%02X%02X%02X' % (r(),r(),r())
        line_colors.append(random_color)

    feature = Feature(geometry=LineString(combine), properties={'stroke' : random_color})
    feature_list.append(feature)

feature_collection = FeatureCollection(feature_list)

import json
with open('trips.geojson', 'w') as f:
    json.dump(feature_collection, f)

```
